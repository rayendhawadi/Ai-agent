```python
# LangChain
from langchain_text_splitters import RecursiveCharacterTextSplitter  #
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage

# LangGraph
from langgraph.graph import StateGraph, END

# Typage de l'état
from typing import TypedDict, Optional

# Lecture PDF
import fitz  # c'est pymupdf qui s'importe sous ce nom

print("✅ Imports OK")
```

```python
# Le LLM — tu connais déjà
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.2   # plus bas que le projet 1 car on veut des résumés factuels
)

# Le modèle d'embeddings — tourne en local sur le CPU de Colab
embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-small-en-v1.5",  # modèle léger, très bon rapport qualité/vitesse
    model_kwargs={"device": "cpu"},         # on utilise le CPU pour les embeddings
    encode_kwargs={"normalize_embeddings": True}  # normalise les vecteurs pour de meilleures recherches
)

print("✅ LLM Groq prêt")
print("✅ Modèle d'embeddings BAAI/bge-small-en-v1.5 chargé")
```

```python
class PDFAgentState(TypedDict):
    pdf_path: str                    # chemin vers le fichier PDF uploadé
    raw_text: str                    # texte brut extrait du PDF
    chunks: list                     # liste des morceaux de texte
    vectorstore: Optional[object]    # la base vectorielle FAISS
    summary: str                     # le résumé généré
    question: str                    # la question posée par l'utilisateur
    answer: str                      # la réponse à la question
    num_pages: int                   # nombre de pages du PDF
    error: str                       # message d'erreur si quelque chose échoue

print("✅ État de l'agent défini")
```

```python
def load_pdf(pdf_path: str) -> dict:
    """
    Lit un fichier PDF et extrait tout son texte page par page.
    Retourne le texte brut et le nombre de pages.
    """
    try:
        doc = fitz.open(pdf_path)        # ouvre le PDF
        num_pages = len(doc)             # compte le nombre de pages
        full_text = ""

        for page_num in range(num_pages):
            page = doc[page_num]                         # accède à la page
            text = page.get_text("text")                 # extrait le texte
            full_text += f"\n--- Page {page_num + 1} ---\n{text}"  # ajoute un marqueur de page

        doc.close()

        print(f"✅ PDF lu avec succès : {num_pages} pages, {len(full_text)} caractères extraits")
        return {"raw_text": full_text, "num_pages": num_pages, "error": ""}

    except Exception as e:
        print(f"❌ Erreur lors de la lecture du PDF : {str(e)}")
        return {"raw_text": "", "num_pages": 0, "error": str(e)}
```

```python
def split_text(raw_text: str) -> list:
    """
    Découpe le texte brut en morceaux (chunks) adaptés au LLM.
    """
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,        # chaque chunk fait max 1000 caractères
        chunk_overlap=150,      # 150 caractères de chevauchement entre chunks
        separators=["\n\n", "\n", ".", " "]  # ordre de priorité pour couper
    )

    chunks = splitter.split_text(raw_text)

    print(f"✅ Texte découpé en {len(chunks)} chunks")
    print(f"   Taille moyenne : {sum(len(c) for c in chunks) // len(chunks)} caractères/chunk")
    return chunks
```

```python
def create_vectorstore(chunks: list) -> object:
    """
    Convertit les chunks en vecteurs et les stocke dans FAISS.
    """
    print("⏳ Création du VectorStore en cours...")

    vectorstore = FAISS.from_texts(
        texts=chunks,
        embedding=embeddings
    )

    print(f"✅ VectorStore créé avec {len(chunks)} vecteurs indexés")
    return vectorstore


def search_vectorstore(vectorstore, query: str, k: int = 4) -> str:
    """
    Cherche les k chunks les plus pertinents par rapport à une requête.
    """
    docs = vectorstore.similarity_search(query, k=k)
    context = "\n\n".join([doc.page_content for doc in docs])
    return context
```

```python
SUMMARIZER_PROMPT = """Tu es un expert en analyse et synthèse de documents professionnels.

Ton rôle : lire le contenu d'un document et produire un résumé professionnel structuré.

Format attendu :
**1. Sujet principal** : (1-2 phrases sur le thème central du document)

**2. Points clés** : (5-7 points essentiels extraits du document)

**3. Données et chiffres importants** : (tous les chiffres, dates, statistiques mentionnés)

**4. Conclusions** : (ce que le document conclut ou recommande)

**5. Mots-clés** : (8-10 mots-clés représentatifs du document)

Règles strictes :
- Reste UNIQUEMENT fidèle au contenu du document, n'invente rien
- Si une information n'est pas dans le document, ne la mentionne pas
- Style : professionnel, concis, objectif
- Longueur totale : 400-500 mots"""


QA_PROMPT = """Tu es un assistant expert en analyse de documents.

Tu dois répondre à la question de l'utilisateur en te basant UNIQUEMENT sur le contexte fourni.

Règles strictes :
- Si la réponse est dans le contexte, réponds de manière précise et complète
- Si la réponse n'est PAS dans le contexte, dis clairement : "Cette information ne figure pas dans le document."
- Ne jamais inventer ou compléter avec des connaissances externes
- Cite la partie du document qui justifie ta réponse
- Style : professionnel, direct, structuré

Longueur : 100-200 mots maximum."""

print("✅ Prompts définis")
```

```python
# ============================================================
# 📥 NŒUD 1 — LOADER : lit le PDF et crée le VectorStore
# ============================================================
def loader_node(state: PDFAgentState):
    print("\n📥 LOADER — Lecture du PDF...")

    # Étape 1 : lire le PDF
    result = load_pdf(state["pdf_path"])

    if result["error"]:
        return {"error": result["error"]}

    # Étape 2 : découper en chunks
    chunks = split_text(result["raw_text"])

    # Étape 3 : créer le VectorStore
    vectorstore = create_vectorstore(chunks)

    return {
        "raw_text":    result["raw_text"],
        "num_pages":   result["num_pages"],
        "chunks":      chunks,
        "vectorstore": vectorstore,
        "error":       ""
    }


# ============================================================
# 📝 NŒUD 2 — SUMMARIZER : génère le résumé structuré
# ============================================================
def summarizer_node(state: PDFAgentState):
    print("\n📝 SUMMARIZER — Génération du résumé...")

    # On prend les 6000 premiers caractères pour le résumé
    # (suffisant pour capturer l'essentiel sans dépasser les tokens)
    text_for_summary = state["raw_text"][:6000]

    msg = llm.invoke([
        SystemMessage(content=SUMMARIZER_PROMPT),
        HumanMessage(content=f"Voici le contenu du document à résumer :\n\n{text_for_summary}")
    ])

    print("✅ Résumé généré")
    return {"summary": msg.content}


# ============================================================
# ❓ NŒUD 3 — QA : répond aux questions sur le document
# ============================================================
def qa_node(state: PDFAgentState):
    print(f"\n❓ QA — Recherche de la réponse à : '{state['question']}'")

    # Recherche les chunks les plus pertinents pour la question
    context = search_vectorstore(state["vectorstore"], state["question"], k=4)

    msg = llm.invoke([
        SystemMessage(content=QA_PROMPT),
        HumanMessage(content=f"Contexte extrait du document :\n{context}\n\nQuestion : {state['question']}")
    ])

    print("✅ Réponse générée")
    return {"answer": msg.content}
```

```python
def route_after_loader(state: PDFAgentState):
    """
    Après le loader, on vérifie si une erreur s'est produite.
    Si oui, on arrête. Sinon, on continue vers le summarizer.
    """
    if state.get("error"):
        print(f"❌ Erreur détectée : {state['error']}")
        return "end"
    return "summarize"


def route_after_summary(state: PDFAgentState):
    """
    Après le résumé, on vérifie si une question a été posée.
    Si oui, on passe au QA. Sinon, on s'arrête.
    """
    if state.get("question") and state["question"].strip():
        return "qa"
    return "end"

print("✅ Routage défini")
```

```python
workflow = StateGraph(PDFAgentState)

# Ajout des nœuds
workflow.add_node("loader",     loader_node)
workflow.add_node("summarizer", summarizer_node)
workflow.add_node("qa",         qa_node)

# Point d'entrée
workflow.set_entry_point("loader")

# Routage conditionnel après le loader
workflow.add_conditional_edges(
    "loader",
    route_after_loader,
    {
        "summarize": "summarizer",
        "end":       END
    }
)

# Routage conditionnel après le summarizer
workflow.add_conditional_edges(
    "summarizer",
    route_after_summary,
    {
        "qa":  "qa",
        "end": END
    }
)

# Le QA termine toujours le pipeline
workflow.add_edge("qa", END)

app = workflow.compile()
print("✅ Graphe compilé avec succès !")
```

```python
from google.colab import files

# Upload du PDF
print("📂 Sélectionne ton fichier PDF...")
uploaded = files.upload()

# Récupère le chemin du fichier uploadé
pdf_filename = list(uploaded.keys())[0]
pdf_path = f"/content/{pdf_filename}"
print(f"✅ Fichier reçu : {pdf_filename}")

# Lancement du pipeline — résumé uniquement
inputs = {
    "pdf_path":    pdf_path,
    "raw_text":    "",
    "chunks":      [],
    "vectorstore": None,
    "summary":     "",
    "question":    "",   # pas de question → juste le résumé
    "answer":      "",
    "num_pages":   0,
    "error":       ""
}

print("\n🚀 Lancement du pipeline...")
result = app.invoke(inputs)

# Affichage du résultat
print("\n" + "═" * 60)
print(f"  📄  RÉSUMÉ DU DOCUMENT ({result['num_pages']} pages)")
print("═" * 60)
print(result["summary"])
```

```python
# Réutilise le vectorstore déjà créé pour poser une question
question = "quel est le nom de la prof qui fait ce cours ?"  # change selon ton PDF

inputs_qa = {
    **result,           # on reprend tout l'état précédent
    "question": question
}

print(f"❓ Question : {question}\n")
result_qa = app.invoke(inputs_qa)

print("═" * 60)
print("  💬  RÉPONSE")
print("═" * 60)
print(result_qa["answer"])
```
