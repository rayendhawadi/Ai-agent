```python
from typing import TypedDict, List
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_community.tools.tavily_search import TavilySearchResults
from langgraph.graph import StateGraph, END
import gradio as gr

print("✅ Imports OK")
```

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.2
)

search_tool = TavilySearchResults(
    max_results=5,
    search_depth="advanced"
)

print("✅ LLM Groq prêt")
print("✅ Outil Tavily prêt")
```

```python
class NewsletterState(TypedDict):
    sujet: str                  # sujet donné par l'utilisateur
    periode: str                # ex: "cette semaine", "ce mois"
    queries: List[str]          # requêtes de recherche générées
    articles_bruts: List[dict]  # résultats bruts de Tavily
    articles_filtres: List[dict]# articles filtrés et classés
    resumes: List[str]          # mini-résumés par article
    bulletin: str               # bulletin final généré
    error: str                  # message d'erreur si échec

print("✅ État défini")
```

```python
QUERY_PROMPT = """Tu es un expert en veille technologique et stratégique.

Ton rôle : transformer un sujet en 4 requêtes de recherche web optimisées pour trouver les actualités les plus récentes.

Instructions :
- Génère exactement 4 requêtes en anglais (meilleurs résultats sur le web)
- Couvre différents angles : news générales, recherche, entreprises, impact
- Inclus toujours l'année en cours dans au moins 2 requêtes
- Format de réponse : une requête par ligne, sans numérotation, sans commentaire

Exemple pour le sujet "blockchain" :
blockchain latest news 2025
blockchain technology breakthroughs this week
top blockchain companies developments April 2025
blockchain real world applications 2025"""


FILTER_PROMPT = """Tu es un expert en analyse et curation de contenu technologique.

Ton rôle : analyser une liste d'articles et sélectionner uniquement les plus pertinents et récents.

Critères de sélection :
- Pertinence directe avec le sujet
- Récence de l'information (priorité aux plus récents)
- Qualité et fiabilité de la source
- Valeur informative (évite les articles trop généraux ou publicitaires)

Format de réponse : retourne uniquement les titres des articles retenus, un par ligne, sans commentaire."""


SUMMARIZER_PROMPT = """Tu es un analyste technologique senior.

Ton rôle : lire un article et produire un résumé factuel de 3-4 phrases en français.

Instructions :
- Capture le fait principal, l'acteur concerné et l'impact potentiel
- Inclus les chiffres et données importantes si présents
- Style : factuel, professionnel, sans opinion personnelle
- Termine par une phrase sur l'impact ou la signification de cette news

Réponds UNIQUEMENT avec le résumé, sans titre ni commentaire."""


WRITER_PROMPT = """Tu es un rédacteur spécialisé en bulletins de veille technologique professionnels.

Ton rôle : assembler des résumés d'articles en un bulletin hebdomadaire structuré et professionnel en français.

Format obligatoire du bulletin :

📰 BULLETIN DE VEILLE — [SUJET EN MAJUSCULES]
📅 Période : [période]
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

🔍 INTRODUCTION
(2-3 phrases sur le contexte général du sujet cette semaine)

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
📌 LES ACTUALITÉS CLÉS
(liste numérotée des news principales, 1 paragraphe par news)

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
📊 CHIFFRES & DONNÉES DE LA SEMAINE
(bullet points avec les données chiffrées importantes)

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
🔮 TENDANCES À SURVEILLER
(2-3 tendances émergentes à garder en tête)

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
✅ CONCLUSION
(3-4 phrases de synthèse)

Règles :
- Reste fidèle aux informations des résumés fournis
- Style : professionnel, clair, sans jargon inutile
- Longueur totale : 600-800 mots"""

print("✅ Prompts définis")
```

```python

def query_generator_node(state: NewsletterState):
    print("\n🔍 QUERY GENERATOR — Génération des requêtes...")

    msg = llm.invoke([
        SystemMessage(content=QUERY_PROMPT),
        HumanMessage(content=f"Sujet : {state['sujet']}\nPériode : {state['periode']}")
    ])


    queries = [q.strip() for q in msg.content.strip().split('\n') if q.strip()]

    print(f"✅ {len(queries)} requêtes générées :")
    for q in queries:
        print(f"   → {q}")

    return {"queries": queries}



def searcher_node(state: NewsletterState):
    print("\n🌐 SEARCHER — Recherche web en cours...")

    all_articles = []

    for query in state["queries"]:
        try:
            results = search_tool.invoke(query)
            for r in results:
                all_articles.append({
                    "titre":   r.get("title", "Sans titre"),
                    "contenu": r.get("content", ""),
                    "url":     r.get("url", ""),
                    "query":   query
                })
            print(f"   ✅ '{query}' → {len(results)} articles trouvés")
        except Exception as e:
            print(f"   ❌ Erreur pour '{query}' : {str(e)}")

    print(f"\n✅ Total : {len(all_articles)} articles collectés")
    return {"articles_bruts": all_articles, "error": ""}



def filter_node(state: NewsletterState):
    print("\n🔎 FILTER — Tri et filtrage des articles...")

    if not state["articles_bruts"]:
        return {"articles_filtres": [], "error": "Aucun article trouvé."}


    articles_text = "\n".join([
        f"- {a['titre']} ({a['url']})"
        for a in state["articles_bruts"]
    ])

    msg = llm.invoke([
        SystemMessage(content=FILTER_PROMPT),
        HumanMessage(content=f"Sujet : {state['sujet']}\n\nArticles disponibles :\n{articles_text}")
    ])


    titres_retenus = [t.strip() for t in msg.content.strip().split('\n') if t.strip()]


    articles_filtres = []
    for article in state["articles_bruts"]:
        for titre in titres_retenus:
            if titre.lower()[:30] in article["titre"].lower():
                articles_filtres.append(article)
                break


    if len(articles_filtres) < 3:
        articles_filtres = state["articles_bruts"][:8]

    print(f"✅ {len(articles_filtres)} articles retenus après filtrage")
    return {"articles_filtres": articles_filtres}



def summarizer_node(state: NewsletterState):
    print("\n📝 SUMMARIZER — Résumé de chaque article...")

    resumes = []

    for i, article in enumerate(state["articles_filtres"]):
        contenu = article["contenu"][:1500]
        try:
            msg = llm.invoke([
                SystemMessage(content=SUMMARIZER_PROMPT),
                HumanMessage(content=f"Titre : {article['titre']}\n\nContenu : {contenu}\n\nSource : {article['url']}")
            ])
            resume = f"**{article['titre']}**\n{msg.content}\n(Source : {article['url']})"
            resumes.append(resume)
            print(f"   ✅ Article {i+1}/{len(state['articles_filtres'])} résumé")
        except Exception as e:
            print(f"   ❌ Erreur article {i+1} : {str(e)}")

    print(f"\n✅ {len(resumes)} résumés générés")
    return {"resumes": resumes}



def writer_node(state: NewsletterState):
    print("\n📰 WRITER — Génération du bulletin final...")

    resumes_text = "\n\n".join(state["resumes"])

    msg = llm.invoke([
        SystemMessage(content=WRITER_PROMPT),
        HumanMessage(content=f"Sujet : {state['sujet']}\nPériode : {state['periode']}\n\nRésumés des articles :\n\n{resumes_text}")
    ])

    print("✅ Bulletin généré !")
    return {"bulletin": msg.content}

print("✅ Nœuds définis")
```

```python
def route_after_search(state: NewsletterState):
    if state.get("error") or not state["articles_bruts"]:
        print("❌ Aucun article trouvé, arrêt du pipeline.")
        return "end"
    return "filter"

print("✅ Routage défini")
```

```python
workflow = StateGraph(NewsletterState)


workflow.add_node("query_generator", query_generator_node)
workflow.add_node("searcher",        searcher_node)
workflow.add_node("filter",          filter_node)
workflow.add_node("summarizer",      summarizer_node)
workflow.add_node("writer",          writer_node)


workflow.set_entry_point("query_generator")


workflow.add_edge("query_generator", "searcher")

workflow.add_conditional_edges(
    "searcher",
    route_after_search,
    {
        "filter": "filter",
        "end":    END
    }
)


workflow.add_edge("filter",    "summarizer")
workflow.add_edge("summarizer","writer")
workflow.add_edge("writer",    END)

app = workflow.compile()
print("✅ Graphe compilé avec succès !")
```

```python
inputs = {
    "sujet":           "intelligence artificielle",
    "periode":         "cette semaine",
    "queries":         [],
    "articles_bruts":  [],
    "articles_filtres":[],
    "resumes":         [],
    "bulletin":        "",
    "error":           ""
}

print("🚀 Lancement du pipeline de veille...\n")
result = app.invoke(inputs)

print("\n" + "═" * 60)
print("  📰  BULLETIN FINAL")
print("═" * 60)
print(result["bulletin"])
```

```python
def run_veille(sujet, periode):
    if not sujet.strip():
        return "❌ Veuillez entrer un sujet.", ""

    inputs = {
        "sujet":            sujet,
        "periode":          periode,
        "queries":          [],
        "articles_bruts":   [],
        "articles_filtres": [],
        "resumes":          [],
        "bulletin":         "",
        "error":            ""
    }

    log = f"🚀 Lancement de la veille sur : **{sujet}**\n"
    log += f"📅 Période : {periode}\n\n"
    log += "⏳ Génération des requêtes...\n"
    log += "⏳ Recherche web en cours...\n"
    log += "⏳ Filtrage et résumés...\n"
    log += "⏳ Rédaction du bulletin...\n"

    result = app.invoke(inputs)

    if result.get("error"):
        return f"❌ Erreur : {result['error']}", ""

    log += f"\n✅ {len(result['articles_filtres'])} articles analysés"
    log += f"\n✅ {len(result['resumes'])} résumés générés"
    log += "\n✅ Bulletin prêt !"

    return log, result["bulletin"]



with gr.Blocks(theme=gr.themes.Soft(), title="📰 Agent Veille Technologique") as demo:

    gr.Markdown("""
    # 📰 Agent de Veille Technologique
    ### Entrez un sujet → l'agent cherche les news → génère un bulletin professionnel en français
    """)

    with gr.Row():
        with gr.Column(scale=2):
            sujet_input = gr.Textbox(
                label="🔎 Sujet de veille",
                placeholder="Ex: intelligence artificielle, blockchain, cybersécurité...",
                lines=1
            )
        with gr.Column(scale=1):
            periode_input = gr.Dropdown(
                label="📅 Période",
                choices=["cette semaine", "ce mois", "aujourd'hui", "ces 3 derniers mois"],
                value="cette semaine"
            )

    run_btn = gr.Button("🚀 Générer le bulletin", variant="primary", size="lg")

    with gr.Row():
        with gr.Column(scale=1):
            log_output = gr.Markdown(
                label="📋 Progression",
                value="En attente..."
            )
        with gr.Column(scale=3):
            bulletin_output = gr.Markdown(
                label="📰 Bulletin généré",
                value="Le bulletin apparaîtra ici..."
            )

    run_btn.click(
        fn=run_veille,
        inputs=[sujet_input, periode_input],
        outputs=[log_output, bulletin_output]
    )

demo.launch(debug=True, share=True)
```
